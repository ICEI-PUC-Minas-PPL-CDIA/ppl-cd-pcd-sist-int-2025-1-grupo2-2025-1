```
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, f1_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
# RFECV e LightGBM são removidos pois não faremos mais a pré-seleção com LGBM para a RNA
import warnings
import re
import pickle
import time
import traceback

# Imports para Ray Tune
import ray
from ray import tune
from ray.tune.schedulers import ASHAScheduler
from ray.tune.search.hyperopt import HyperOptSearch # Requer: pip install hyperopt

# Imports para Keras (TensorFlow)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout, concatenate, BatchNormalization, Activation
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import regularizers # Para L2
from ray.train.tensorflow.keras import ReportCheckpointCallback

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

os.makedirs('visualizacoes_classificacao_salario_v7_nn_completo_ray_tune', exist_ok=True)

# --- Funções de limpeza de dados (sem alterações) ---
def clean_col_name(col_name):
    original_input = col_name
    if isinstance(col_name, tuple): col_name = "_".join(str(item).strip() for item in col_name)
    elif not isinstance(col_name, str): col_name = str(col_name)
    col_name = re.sub(r'[^\w\s-]', '', col_name).strip()
    col_name = re.sub(r'[-\s]+', '_', col_name); col_name = re.sub(r"_+", "_", col_name)
    col_name = col_name.strip("_");
    if not col_name: return f"col_limpa_vazia_{hash(original_input)}"
    if col_name and col_name[0].isdigit(): col_name = "_" + col_name
    return col_name

def extract_salary_lower_bound(salary_range_str):
    if pd.isna(salary_range_str): return np.nan
    s = str(salary_range_str).lower().replace('r$', '').replace('.', '').replace('/mês', '').strip()
    match_de_a = re.search(r'de\s*(\d+)\s*a\s*(\d+)', s)
    if match_de_a: return float(match_de_a.group(1))
    match_acima_de = re.search(r'acima de\s*(\d+)', s)
    if match_acima_de: return float(match_acima_de.group(1))
    match_menos_de = re.search(r'menos de\s*(\d+)', s)
    if match_menos_de: return 0
    match_so_numeros = re.findall(r'\d+', s)
    if match_so_numeros: return float(match_so_numeros[0])
    return np.nan

def map_uf_to_region(uf_series: pd.Series) -> pd.Series:
    mapa_regioes = {
        'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste',
        'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
        'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
        'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
        'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
        'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
    }
    uf_series_normalized = uf_series.astype(str).str.upper().str.strip()
    siglas_uf = list(mapa_regioes.keys())
    def extract_sigla(val):
        if val in siglas_uf: return val
        for sigla in siglas_uf:
            if f"({sigla})" in val or f" {sigla} " in val or val.endswith(f" {sigla}"): return sigla
        if "DISTRITO FEDERAL" in val: return "DF"
        if "SAO PAULO" in val: return "SP"
        if "RIO DE JANEIRO" in val: return "RJ"
        if "MINAS GERAIS" in val: return "MG"
        if "ESPIRITO SANTO" in val: return "ES"
        if "RIO GRANDE DO SUL" in val: return "RS"
        if "SANTA CATARINA" in val: return "SC"
        if "PARANA" in val: return "PR"
        return val
    uf_series_normalized = uf_series_normalized.apply(extract_sigla)
    mapped_series = uf_series_normalized.map(mapa_regioes)
    return mapped_series.fillna('Desconhecida')

def clean_experience_to_numeric(exp_val):
    if pd.isna(exp_val): return np.nan
    s = str(exp_val).lower().strip()
    if 'menos de 1 ano' in s or 'menos de um ano' in s or '< 1 ano' in s: return 0.5
    if 'não tenho experiência' in s or 'sem experiência' in s: return 0
    numbers = re.findall(r'\d+\.?\d*', s)
    if numbers: return float(numbers[0])
    return np.nan
# --- [FIM DAS Funções de limpeza de dados] ---


# Função para criar o modelo Keras com Embeddings, Batch Norm e L2
def create_keras_model_v2(embedding_infos_list, num_numerical_features, config_nn):
    inputs = []
    embedded_layers = []
    l2_strength_emb = config_nn.get("l2_strength_embedding", 0.0001) # Hiperparâmetro para L2 nas embeddings

    for emb_info in embedding_infos_list:
        input_cat = Input(shape=(emb_info['input_length'],), name=f"input_{emb_info['name']}")
        inputs.append(input_cat)
        embedding_output_dim = config_nn.get(f"emb_dim_{emb_info['name']}", emb_info['output_dim'])
        embedding = Embedding(input_dim=emb_info['input_dim'],
                              output_dim=embedding_output_dim,
                              name=f"embedding_{emb_info['name']}",
                              embeddings_regularizer=regularizers.l2(l2_strength_emb))(input_cat) # L2 nas Embeddings
        # Batch Normalization pode ser aplicada após embeddings, antes do Flatten
        # embedding_bn = BatchNormalization(name=f"bn_emb_{emb_info['name']}")(embedding)
        # flattened_embedding = Flatten(name=f"flatten_{emb_info['name']}")(embedding_bn)
        flattened_embedding = Flatten(name=f"flatten_{emb_info['name']}")(embedding) # Simples por enquanto
        embedded_layers.append(flattened_embedding)

    if num_numerical_features > 0:
        input_num = Input(shape=(num_numerical_features,), name="input_numerical")
        inputs.append(input_num)
        to_concatenate = embedded_layers + [input_num]
    elif embedded_layers:
        to_concatenate = embedded_layers
    else: raise ValueError("O modelo precisa de pelo menos uma feature de entrada.")

    if len(to_concatenate) > 1: merged_features = concatenate(to_concatenate)
    elif to_concatenate: merged_features = to_concatenate[0]
    else: raise ValueError("Lista de features para concatenar está vazia.")

    l2_strength_dense = config_nn.get("l2_strength_dense", 0.001) # Hiperparâmetro para L2 nas densas

    x = Dense(config_nn.get("dense_units_1", 128), kernel_regularizer=regularizers.l2(l2_strength_dense))(merged_features)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = Dropout(config_nn.get("dropout_1", 0.3))(x)

    if config_nn.get("num_hidden_layers", 1) > 1:
        x = Dense(config_nn.get("dense_units_2", 64), kernel_regularizer=regularizers.l2(l2_strength_dense))(x)
        x = BatchNormalization()(x)
        x = Activation("relu")(x)
        x = Dropout(config_nn.get("dropout_2", 0.3))(x)

    output = Dense(1, activation="sigmoid", name="output")(x)
    model = Model(inputs=inputs, outputs=output)

    optimizer_type = config_nn.get("optimizer", "adam")
    learning_rate = config_nn.get("learning_rate_nn", 0.001)
    if optimizer_type == "adamw":
        optimizer = tf.keras.optimizers.AdamW(learning_rate=learning_rate, weight_decay=config_nn.get("weight_decay", 0.004))
    elif optimizer_type == "nadam":
        optimizer = tf.keras.optimizers.Nadam(learning_rate=learning_rate)
    else: # Default to Adam
        optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])
    return model

# Função objetivo para Ray Tune com Keras (CALLBACK ATUALIZADO)
def objective_ray_tune_keras_v2(config_keras, X_train_list, y_train_arr, X_val_list, y_val_arr,
                             embedding_infos_obj, num_numerical_features_obj):
    tf.keras.backend.clear_session() # Limpar sessão Keras
    model = create_keras_model_v2(embedding_infos_obj, num_numerical_features_obj, config_keras)

    # Callbacks
    early_stopping_patience = config_keras.get("early_stopping_patience", 10)
    early_stopping = EarlyStopping(monitor='val_accuracy', patience=early_stopping_patience,
                                   restore_best_weights=True, mode='max')
    report_checkpoint_callback = ReportCheckpointCallback(metrics={"val_accuracy_tune": "val_accuracy"})

    # Learning Rate Scheduler (exemplo: ReduceLROnPlateau)
    reduce_lr_patience = config_keras.get("reduce_lr_patience", 3)
    reduce_lr_factor = config_keras.get("reduce_lr_factor", 0.2)
    reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=reduce_lr_factor,
                                                              patience=reduce_lr_patience, min_lr=1e-6, verbose=0)

    model.fit(
        X_train_list, y_train_arr,
        validation_data=(X_val_list, y_val_arr),
        epochs=config_keras.get("epochs", 50),
        batch_size=config_keras.get("batch_size", 32),
        callbacks=[early_stopping, report_checkpoint_callback, reduce_lr_callback],
        verbose=0
    )
    # O ReportCheckpointCallback já reporta a métrica para o Ray Tune.

def train_classification_model_salary_range_v7_final_nn_v2(file_path, n_ray_tune_samples_nn=50, ray_tune_timeout_nn=3600):
    if not ray.is_initialized():
        num_cpus_to_use = os.cpu_count();
        if num_cpus_to_use is None or num_cpus_to_use < 1: num_cpus_to_use = 2
        print(f"Inicializando Ray com {num_cpus_to_use} CPUs...")
        ray.init(ignore_reinit_error=True, num_cpus=num_cpus_to_use)

    try:
        df_main = pd.read_excel(file_path)
        print(f"Base de dados carregada: {df_main.shape[0]} linhas, {df_main.shape[1]} colunas")
        original_columns = df_main.columns.tolist()
        df_main.columns = [clean_col_name(col) for col in original_columns]

        col_mapping = {}
        # ... (seu mapeamento de colunas aqui - COPIADO DO CÓDIGO ANTERIOR)
        target_mapped, faixa_etaria_mapped, genero_mapped, nivel_ensino_mapped = False, False, False, False
        tempo_experiencia_p2i_mapped, senioridade_p2g_mapped, uf_mora_p1i1_mapped, cargo_atual_p2f_mapped = False, False, False, False
        for col_idx, col_clean_name in enumerate(df_main.columns):
            if not target_mapped and ("P2_h" in col_clean_name or "faixa_salarial" in col_clean_name.lower()):
                col_mapping["target_original_salary_range"] = col_clean_name; target_mapped = True
            elif not tempo_experiencia_p2i_mapped and ("P2_i" in col_clean_name):
                col_mapping["tempo_experiencia_P2i"] = col_clean_name; tempo_experiencia_p2i_mapped = True
            elif not senioridade_p2g_mapped and ("P2_g" in col_clean_name):
                col_mapping["nivel_senioridade_P2g"] = col_clean_name; senioridade_p2g_mapped = True
            elif not uf_mora_p1i1_mapped and ("P1_i_1" in col_clean_name):
                col_mapping["uf_mora_P1i1"] = col_clean_name; uf_mora_p1i1_mapped = True
            elif not cargo_atual_p2f_mapped and ("P2_f" in col_clean_name):
                col_mapping["cargo_atual_P2f"] = col_clean_name; cargo_atual_p2f_mapped = True
            elif not faixa_etaria_mapped and ("P1_a_1" in col_clean_name or "faixa_etaria" in col_clean_name.lower()):
                col_mapping["faixa_etaria"] = col_clean_name; faixa_etaria_mapped = True
            elif not genero_mapped and ("P1_b" in col_clean_name or "genero" in col_clean_name.lower()):
                col_mapping["genero"] = col_clean_name; genero_mapped = True
            elif not nivel_ensino_mapped and ("P1_l" in col_clean_name or "nivel_ensino" in col_clean_name.lower()):
                col_mapping["nivel_ensino"] = col_clean_name; nivel_ensino_mapped = True
        target_col_agrupada_name = "faixa_salarial_agrupada_2cat_v7_rfecv_fixed_cut" # Manter nome para consistência se quiser
        col_mapping["target"] = target_col_agrupada_name

        # ... (pré-processamento inicial e criação da variável alvo - COPIADO DO CÓDIGO ANTERIOR)
        if "target_original_salary_range" not in col_mapping: print(f"Erro Crítico: Coluna original da faixa salarial não mapeada."); return None
        if "tempo_experiencia_P2i" in col_mapping:
            exp_col_name = col_mapping["tempo_experiencia_P2i"]
            if exp_col_name in df_main.columns: df_main[exp_col_name] = df_main[exp_col_name].apply(clean_experience_to_numeric)
        original_salary_col_name = col_mapping["target_original_salary_range"]
        df_main_processed = df_main.dropna(subset=[original_salary_col_name]).copy()
        df_main_processed['salary_numeric_lower_bound'] = df_main_processed[original_salary_col_name].apply(extract_salary_lower_bound)
        df_main_processed.dropna(subset=['salary_numeric_lower_bound'], inplace=True)
        if df_main_processed.empty: print("Erro: DataFrame vazio após remover NaNs de 'salary_numeric_lower_bound'."); return None
        salary_group_labels = ["Salário Baixo", "Salário Alto"]
        point_of_cut_fixed = 7500.0
        min_salary = df_main_processed['salary_numeric_lower_bound'].min()
        max_salary = df_main_processed['salary_numeric_lower_bound'].max()
        final_bins = []; final_labels = []
        if point_of_cut_fixed <= min_salary: final_bins = [min_salary, max_salary]; final_labels = [salary_group_labels[1]]
        elif point_of_cut_fixed >= max_salary: final_bins = [min_salary, max_salary]; final_labels = [salary_group_labels[0]]
        else: final_bins = [min_salary, point_of_cut_fixed, max_salary]; final_labels = salary_group_labels
        unique_sorted_bins = sorted(list(set(final_bins)))
        try:
            df_main_processed[target_col_agrupada_name] = pd.cut(df_main_processed['salary_numeric_lower_bound'], bins=unique_sorted_bins,
                labels=final_labels if len(unique_sorted_bins) == len(final_labels) + 1 else final_labels[:len(unique_sorted_bins)-1],
                include_lowest=True, duplicates='drop')
            if len(df_main_processed[target_col_agrupada_name].cat.categories) < 2 and len(final_labels) == 2 :
                 if point_of_cut_fixed <= min_salary + 1e-9: df_main_processed[target_col_agrupada_name] = salary_group_labels[1]
                 elif point_of_cut_fixed >= max_salary - 1e-9: df_main_processed[target_col_agrupada_name] = salary_group_labels[0]
                 else: raise ValueError("pd.cut resultou em 1 categoria, forçando fallback.")
        except ValueError:
             print("INFO: pd.cut falhou, usando fallback pd.qcut.")
             df_main_processed[target_col_agrupada_name] = pd.qcut(df_main_processed['salary_numeric_lower_bound'], q=2, labels=salary_group_labels, duplicates='drop')
             if len(df_main_processed[target_col_agrupada_name].cat.categories) < 2:
                 df_main_processed[target_col_agrupada_name] = salary_group_labels[0]
        df_main_processed.dropna(subset=[target_col_agrupada_name], inplace=True)
        print(f"Contagens da variável alvo '{target_col_agrupada_name}':\n{df_main_processed[target_col_agrupada_name].value_counts(dropna=False)}")

        # --- FEATURES PARA A RNA (SEM RFECV do LGBM) ---
        # Usaremos todas as features iniciais que foram mapeadas e são relevantes
        features_para_nn_originais = [
            "faixa_etaria", "genero", "nivel_ensino",
            "tempo_experiencia_P2i", "nivel_senioridade_P2g",
            "uf_mora_P1i1", # Será convertido para Regiao_Mapeada
            "cargo_atual_P2f"
        ]
        feature_cols_nn = [col_mapping[col_internal] for col_internal in features_para_nn_originais if col_internal in col_mapping and col_mapping[col_internal] in df_main_processed.columns]
        X_initial_nn = df_main_processed[feature_cols_nn].copy()

        uf_col_original_name_mapped = col_mapping.get("uf_mora_P1i1")
        if uf_col_original_name_mapped and uf_col_original_name_mapped in X_initial_nn.columns:
            X_initial_nn['Regiao_Mapeada'] = map_uf_to_region(X_initial_nn[uf_col_original_name_mapped])
            X_initial_nn.drop(columns=[uf_col_original_name_mapped], inplace=True)

        print(f"Features consideradas para a RNA (após mapeamento de região): {X_initial_nn.columns.tolist()}")

        le_target = LabelEncoder()
        y_full_nn = pd.Series(le_target.fit_transform(df_main_processed[target_col_agrupada_name]), index=df_main_processed.index)
        if len(le_target.classes_) < 2: print(f"ERRO CRÍTICO: Variável alvo com menos de 2 classes: {le_target.classes_}"); return None

        # Tratamento de Outliers (APENAS em features numéricas que vão para a RNA)
        numerical_cols_for_nn_outlier_treatment = X_initial_nn.select_dtypes(include=np.number).columns.tolist()
        if numerical_cols_for_nn_outlier_treatment:
            outliers_indices_nn = pd.Series(False, index=X_initial_nn.index)
            for col in numerical_cols_for_nn_outlier_treatment:
                Q1 = X_initial_nn[col].quantile(0.25); Q3 = X_initial_nn[col].quantile(0.75); IQR = Q3 - Q1
                lim_inf = Q1 - 1.5 * IQR; lim_sup = Q3 + 1.5 * IQR
                col_outliers = (X_initial_nn[col] < lim_inf) | (X_initial_nn[col] > lim_sup)
                outliers_indices_nn = outliers_indices_nn | col_outliers
            if outliers_indices_nn.sum() > 0:
                print(f"RNA: Removendo {outliers_indices_nn.sum()} outliers.")
                X_initial_nn = X_initial_nn[~outliers_indices_nn]; y_full_nn = y_full_nn[~outliers_indices_nn]
                if X_initial_nn.empty: print("ERRO: Todos os dados removidos como outliers antes do split para RNA."); return None

        # Split dos dados para RNA (Treino HPO e Teste Final)
        X_train_nn_full, X_test_nn_full, y_train_nn_full, y_test_nn_full = train_test_split(
            X_initial_nn, y_full_nn, test_size=0.25, random_state=42, stratify=y_full_nn
        )

        # --- PREPARAÇÃO DOS DADOS PARA REDE NEURAL (Embeddings) ---
        print("\n--- Preparando dados para Rede Neural ---")
        numerical_features_nn = X_train_nn_full.select_dtypes(include=np.number).columns.tolist()
        categorical_features_nn = X_train_nn_full.select_dtypes(exclude=np.number).columns.tolist()
        print(f"Features numéricas para NN: {numerical_features_nn}")
        print(f"Features categóricas para NN: {categorical_features_nn}")

        scaler_nn_final = StandardScaler()
        if numerical_features_nn:
            X_train_numerical_scaled = scaler_nn_final.fit_transform(X_train_nn_full[numerical_features_nn])
            X_test_numerical_scaled = scaler_nn_final.transform(X_test_nn_full[numerical_features_nn])
        else:
            X_train_numerical_scaled = np.array([]).reshape(len(X_train_nn_full), 0)
            X_test_numerical_scaled = np.array([]).reshape(len(X_test_nn_full), 0)

        label_encoders_nn = {}
        X_train_categorical_processed_list = []
        X_test_categorical_processed_list = []
        embedding_infos_final = []
        unk_category_map = {}

        for col in categorical_features_nn:
            le = LabelEncoder()
            X_train_nn_full[col] = X_train_nn_full[col].astype(str).fillna("Missing_Val_Cat_NN")
            X_test_nn_full[col] = X_test_nn_full[col].astype(str).fillna("Missing_Val_Cat_NN")
            le.fit(X_train_nn_full[col])
            train_encoded = le.transform(X_train_nn_full[col])
            unk_idx_for_col = len(le.classes_)
            unk_category_map[col] = unk_idx_for_col
            test_encoded = []
            for item in X_test_nn_full[col]:
                if item in le.classes_: test_encoded.append(le.transform([item])[0])
                else: test_encoded.append(unk_idx_for_col)
            test_encoded = np.array(test_encoded)
            X_train_categorical_processed_list.append(train_encoded.reshape(-1, 1))
            X_test_categorical_processed_list.append(test_encoded.reshape(-1, 1))
            cardinality = len(le.classes_) + 1
            embedding_size = min(50, int(np.ceil(cardinality**0.25) * 4) if cardinality > 1 else 5)
            embedding_infos_final.append({'name': col, 'input_dim': cardinality, 'output_dim': embedding_size, 'input_length': 1})
            label_encoders_nn[col] = le

        X_train_nn_inputs_final_list = X_train_categorical_processed_list + ([X_train_numerical_scaled] if numerical_features_nn else [])
        X_test_nn_inputs_final_list = X_test_categorical_processed_list + ([X_test_numerical_scaled] if numerical_features_nn else [])
        y_train_nn_final_arr = y_train_nn_full.values # Usado para HPO e treino final
        y_test_nn_final_arr = y_test_nn_full.values

        temp_indices = np.arange(len(y_train_nn_final_arr))
        stratify_hpo_y = y_train_nn_final_arr if y_train_nn_final_arr.ndim == 1 and pd.Series(y_train_nn_final_arr).nunique() > 1 and pd.Series(y_train_nn_final_arr).value_counts().min() >=2 else None
        train_hpo_idx, val_hpo_idx = train_test_split(temp_indices, test_size=0.2, random_state=42, stratify=stratify_hpo_y)

        X_train_hpo_nn_list_for_tune = [data[train_hpo_idx] for data in X_train_nn_inputs_final_list]
        y_train_hpo_nn_arr_for_tune = y_train_nn_final_arr[train_hpo_idx]
        X_val_hpo_nn_list_for_tune = [data[val_hpo_idx] for data in X_train_nn_inputs_final_list]
        y_val_hpo_nn_arr_for_tune = y_train_nn_final_arr[val_hpo_idx]

        search_space_nn_ray = {
            "dense_units_1": tune.choice([64, 128, 256, 512]),
            "dense_units_2": tune.choice([32, 64, 128, 256]),
            "dropout_1": tune.quniform(0.1, 0.5, 0.05),
            "dropout_2": tune.quniform(0.1, 0.5, 0.05),
            "learning_rate_nn": tune.loguniform(1e-4, 5e-3), # Ampliado um pouco
            "batch_size": tune.choice([32, 64, 128, 256]),
            "epochs": tune.choice([50, 75, 100]), # ES controlará
            "num_hidden_layers": tune.choice([1, 2, 3]), # Testar com 3 camadas
            "early_stopping_patience": tune.choice([7, 10, 15]),
            "l2_strength_embedding": tune.loguniform(1e-5, 1e-2),
            "l2_strength_dense": tune.loguniform(1e-5, 1e-2),
            "optimizer": tune.choice(["adam", "nadam", "adamw"]),
            "weight_decay": tune.loguniform(1e-5, 1e-2), # Apenas para AdamW
            "reduce_lr_patience": tune.choice([2, 3, 5]),
            "reduce_lr_factor": tune.choice([0.1, 0.2, 0.5])
        }
        for emb_info in embedding_infos_final:
            lower_heuristic_dim = max(4, int(np.ceil(emb_info['input_dim']**0.25) * 2))
            original_heuristic_dim = emb_info['output_dim']
            options_for_emb_dim = sorted(list(set([lower_heuristic_dim, original_heuristic_dim, min(50, emb_info['input_dim']//2 if emb_info['input_dim']//2 >=4 else 4 ) ])))
            if len(options_for_emb_dim) == 1 and options_for_emb_dim[0] < 4: options_for_emb_dim = [4, 8]
            elif options_for_emb_dim and options_for_emb_dim[0] < 4:
                 options_for_emb_dim[0] = 4
                 options_for_emb_dim = sorted(list(set(options_for_emb_dim)))
            if not options_for_emb_dim: options_for_emb_dim = [max(4, original_heuristic_dim // 2), original_heuristic_dim]
            search_space_nn_ray[f"emb_dim_{emb_info['name']}"] = tune.choice(options_for_emb_dim)

        metric_to_optimize_nn = "val_accuracy_tune"
        asha_scheduler = ASHAScheduler(
            time_attr='training_iteration', metric=metric_to_optimize_nn, mode='max',
            max_t=max(search_space_nn_ray["epochs"].categories), grace_period=15, reduction_factor=2 # Mais grace period
        )
        search_algo_nn = HyperOptSearch(metric=metric_to_optimize_nn, mode="max")

        print(f"\nIniciando otimização da RNA com Ray Tune ({n_ray_tune_samples_nn} samples)...")
        analysis_nn = tune.run(
            tune.with_parameters(
                objective_ray_tune_keras_v2,
                X_train_list=X_train_hpo_nn_list_for_tune, y_train_arr=y_train_hpo_nn_arr_for_tune,
                X_val_list=X_val_hpo_nn_list_for_tune, y_val_arr=y_val_hpo_nn_arr_for_tune,
                embedding_infos_obj=embedding_infos_final,
                num_numerical_features_obj=X_train_numerical_scaled.shape[1]
            ),
            config=search_space_nn_ray,
            num_samples=n_ray_tune_samples_nn,
            scheduler=asha_scheduler,
            search_alg=search_algo_nn,
            resources_per_trial={"cpu": 1, "gpu": 0},
            storage_path=os.path.abspath("./ray_results_salary_v7_nn_completo"),
            name=f"Keras_Salary_Clf_RayTune_v7_completo_samples{n_ray_tune_samples_nn}",
            verbose=1, raise_on_failed_trial=False,
            time_budget_s=ray_tune_timeout_nn
        )

        print("\nOtimização da RNA com Ray Tune concluída!")
        best_trial_nn_ray = analysis_nn.get_best_trial(metric=metric_to_optimize_nn, mode="max", scope="all")

        if best_trial_nn_ray and best_trial_nn_ray.last_result :
            best_nn_hpo_score = best_trial_nn_ray.last_result.get(metric_to_optimize_nn, -1.0)
            best_nn_params_hpo = best_trial_nn_ray.config
            print(f"Melhor Acurácia na Validação (HPO da RNA): {best_nn_hpo_score:.4f}")
            print(f"Melhores hiperparâmetros (RNA Ray Tune): {best_nn_params_hpo}")
        else:
            print("ERRO/AVISO: Nenhum trial do Ray Tune para RNA foi bem-sucedido. Usando config default.")
            best_nn_hpo_score = -1.0
            best_nn_params_hpo = { # Configuração default mais robusta
                "dense_units_1": 128, "dense_units_2": 64, "dropout_1": 0.3, "dropout_2": 0.3,
                "learning_rate_nn": 0.001, "batch_size": 64, "epochs": 75, "num_hidden_layers": 2,
                "early_stopping_patience": 10, "l2_strength_embedding": 0.0001, "l2_strength_dense": 0.001,
                "optimizer": "adam", "weight_decay": 0.004, "reduce_lr_patience": 3, "reduce_lr_factor": 0.2
            }
            for emb_info in embedding_infos_final:
                 best_nn_params_hpo[f"emb_dim_{emb_info['name']}"] = emb_info['output_dim']

        print("\nTreinando o modelo RNA final com os melhores hiperparâmetros...")
        # Para o treino final, usamos o conjunto de treino HPO completo (X_train_nn_inputs_final_list, y_train_nn_final_arr)
        # e criamos um novo split de validação para EarlyStopping do treino final.
        final_train_indices, final_val_indices = train_test_split(
            np.arange(len(y_train_nn_final_arr)), test_size=0.15, random_state=123, # 15% para validação final
            stratify=y_train_nn_final_arr if y_train_nn_final_arr.ndim == 1 and pd.Series(y_train_nn_final_arr).nunique() > 1 and pd.Series(y_train_nn_final_arr).value_counts().min() >=2 else None
        )
        X_final_train_list = [data[final_train_indices] for data in X_train_nn_inputs_final_list]
        y_final_train_arr = y_train_nn_final_arr[final_train_indices]
        X_final_val_list = [data[final_val_indices] for data in X_train_nn_inputs_final_list]
        y_final_val_arr = y_train_nn_final_arr[final_val_indices]

        final_nn_model = create_keras_model_v2(embedding_infos_final, X_train_numerical_scaled.shape[1], best_nn_params_hpo)
        final_early_stopping_patience = best_nn_params_hpo.get("early_stopping_patience", 15) + 10 # Mais paciência
        final_early_stopping = EarlyStopping(monitor='val_accuracy', patience=final_early_stopping_patience, restore_best_weights=True, mode='max', verbose=1)
        final_reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss', factor=best_nn_params_hpo.get("reduce_lr_factor",0.2),
            patience=best_nn_params_hpo.get("reduce_lr_patience",3) + 2, # Mais paciência para LR reduce
            min_lr=1e-7, verbose=1
        )
        final_epochs = best_nn_params_hpo.get("epochs", 100) * 2 # Potencialmente mais épocas, ES decide

        print(f"Treinando modelo final RNA em {len(y_final_train_arr)} amostras, validando em {len(y_final_val_arr)} amostras.")
        final_nn_model.fit(
            X_final_train_list, y_final_train_arr,
            validation_data=(X_final_val_list, y_final_val_arr),
            epochs=final_epochs,
            batch_size=best_nn_params_hpo.get("batch_size", 64),
            callbacks=[final_early_stopping, final_reduce_lr_callback],
            verbose=1
        )

        print("\nAvaliando o modelo RNA final no conjunto de teste...")
        loss_test_nn, accuracy_test_nn = final_nn_model.evaluate(X_test_nn_inputs_final_list, y_test_nn_final_arr, verbose=0)
        y_pred_proba_test_nn = final_nn_model.predict(X_test_nn_inputs_final_list)
        y_pred_test_nn = (y_pred_proba_test_nn > 0.5).astype(int)

        f1_test_nn = f1_score(y_test_nn_final_arr, y_pred_test_nn, average='weighted')
        roc_auc_test_nn = roc_auc_score(y_test_nn_final_arr, y_pred_proba_test_nn) if len(np.unique(y_test_nn_final_arr)) > 1 else 0.5
        report_str_test_nn = classification_report(y_test_nn_final_arr, y_pred_test_nn, target_names=le_target.classes_, zero_division=0)
        conf_matrix_test_nn = confusion_matrix(y_test_nn_final_arr, y_pred_test_nn)

        print(f"\n--- Resultados da Avaliação da REDE NEURAL (v2) no CONJUNTO DE TESTE ---")
        print(f"Acurácia no Teste (RNA): {accuracy_test_nn:.4f}")
        print(f"F1-Score (Ponderado) no Teste (RNA): {f1_test_nn:.4f}")
        print(f"ROC AUC no Teste (RNA): {roc_auc_test_nn:.4f}")
        print("\nRelatório de Classificação (Teste - RNA):\n", report_str_test_nn)
        plt.figure(figsize=(8,6))
        sns.heatmap(conf_matrix_test_nn, annot=True, fmt="d", cmap="Blues", xticklabels=le_target.classes_, yticklabels=le_target.classes_)
        plt.title('Matriz de Confusão (Teste - RNA v2)')
        plt.savefig('visualizacoes_classificacao_salario_v7_nn_completo_ray_tune/matriz_confusao_nn_v2_teste.png')
        plt.show(); plt.close()

        model_nn_filename_keras = 'modelo_keras_classificacao_faixa_salarial_v7_nn_completo_raytune.keras'
        final_nn_model.save(model_nn_filename_keras)
        preprocessors_nn_filename_pkl = 'preprocessors_nn_v7_completo_raytune.pkl'
        with open(preprocessors_nn_filename_pkl, 'wb') as f:
            pickle.dump({
                'label_encoders_nn': label_encoders_nn, 'scaler_nn_final': scaler_nn_final,
                'embedding_infos_final': embedding_infos_final,
                'feature_column_names_in_order': X_initial_nn.columns.tolist(), # Ordem das colunas antes do split num/cat
                'numerical_features_nn': numerical_features_nn, 'categorical_features_nn': categorical_features_nn,
                'label_encoder_target': le_target, 'unk_category_map': unk_category_map
            }, f)
        print(f"Modelo Keras salvo em {model_nn_filename_keras}")
        print(f"Pré-processadores da RNA salvos em {preprocessors_nn_filename_pkl}")

        return {
            'nn_accuracy_test': accuracy_test_nn, 'nn_f1_score_test': f1_test_nn, 'nn_roc_auc_test': roc_auc_test_nn,
            'nn_best_hpo_val_accuracy': best_nn_hpo_score, 'nn_best_params': best_nn_params_hpo,
            'nn_model_path': model_nn_filename_keras, 'nn_preprocessors_path': preprocessors_nn_filename_pkl
        }
    except Exception as e:
        print(f"Ocorreu um erro inesperado na função principal da RNA: {e}")
        traceback.print_exc()
        return None
    finally:
        if ray.is_initialized():
           print("Desligando Ray ao final da função RNA...")
           ray.shutdown()

if __name__ == "__main__":
    file_path = "Main_database (2).xlsx"
    if os.path.exists(file_path):
        results_nn = train_classification_model_salary_range_v7_final_nn_v2(
            file_path,
            n_ray_tune_samples_nn=75, # Aumentar um pouco os trials para HPO mais completo
            ray_tune_timeout_nn=5400, # 1.5 horas para HPO
        )
        if results_nn:
            print("\n\n--- Resumo Final da Execução (REDE NEURAL v2 com RayTune) ---")
            print(f"Melhor Acurácia na Validação (HPO da RNA): {results_nn['nn_best_hpo_val_accuracy']:.4f}")
            print(f"Acurácia no Teste (RNA): {results_nn['nn_accuracy_test']:.4f}")
            print(f"F1-Score no Teste (RNA): {results_nn['nn_f1_score_test']:.4f}")
            print(f"ROC AUC no Teste (RNA): {results_nn['nn_roc_auc_test']:.4f}")
            print(f"Melhores Parâmetros (RNA RayTune): {results_nn['nn_best_params']}")
            print(f"Modelo Keras salvo em: {results_nn['nn_model_path']}")
            print(f"Pré-processadores da RNA salvos em: {results_nn['nn_preprocessors_path']}")
    else:
        print(f"ERRO: Arquivo '{file_path}' não encontrado.")
```
