```python
# -*- coding: utf-8 -*-
"""
Passo_a_Passo_GBM_Disparidade_Salarial_Estilo_Iris.ipynb

Adaptado para incluir visualiza√ß√µes de √°rvores individuais.
"""

# --- Etapa 1: Configura√ß√£o do Ambiente, Upload e Carregamento ---

# Importar bibliotecas necess√°rias
import pandas as pd
import numpy as np
import re
import io
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import lightgbm as lgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns

# Adicionar bibliotecas para visualiza√ß√£o de √°rvore estilo Iris
import graphviz # Para plot_tree estilo graphviz
# !pip install dtreeviz # Descomente se n√£o estiver instalado
# from dtreeviz.models.lightgbm_decision_tree import * # Import espec√≠fico para LGBM se dispon√≠vel/necess√°rio
# from dtreeviz import model as dtreeviz_model # Import geral


# --- Upload dos Arquivos ---
print("Por favor, fa√ßa o upload do arquivo 'survey_cleaned.csv':")
uploaded_survey = files.upload()
survey_filename = list(uploaded_survey.keys())[0]
print(f"\nArquivo '{survey_filename}' carregado.")

print("\nPor favor, fa√ßa o upload do arquivo 'microdados_agrupados_por_uf-1.csv' ou similar:")
uploaded_microdados = files.upload()
if uploaded_microdados:
    microdados_filename = list(uploaded_microdados.keys())[0]
    print(f"\nArquivo '{microdados_filename}' carregado.")
else:
    print("\nNenhum arquivo microdados carregado.")
    df_microdados = None

# --- Carregamento dos Dados ---
try:
    df = pd.read_csv(io.BytesIO(uploaded_survey[survey_filename]), sep=',', encoding='utf-8', engine='python')
    print(f"\nDataset Survey carregado. Shape: {df.shape}")
except Exception as e: print(f"Erro ao carregar Survey: {e}"); raise

if uploaded_microdados:
    try:
        df_microdados = pd.read_csv(io.BytesIO(uploaded_microdados[microdados_filename]), sep=',', encoding='utf-8', engine='python')
        print(f"Dataset Microdados carregado. Shape: {df_microdados.shape}")
    except Exception as e: print(f"Erro ao carregar Microdados: {e}"); df_microdados = None

# --- Fun√ß√£o para Limpar Nomes das Colunas ---
def clean_col_names(df_to_clean):
    # ... (fun√ß√£o clean_col_names da vers√£o anterior - v3) ...
    new_cols = {}
    if df_to_clean is None: return None
    for col in df_to_clean.columns:
        col_str = str(col)
        match = re.match(r"\(\s*'(.+?)'\s*,\s*'(.+?)'\s*\)", col_str)
        if match:
            code = match.group(1).strip()
            desc = match.group(2).strip()
            clean_desc = re.sub(r'[?()/,.:#]+', '', desc)
            clean_desc = re.sub(r'\s+', '_', clean_desc.strip())
            clean_desc = re.sub(r'_+', '_', clean_desc)
            clean_desc = re.sub(r'[^\w_]+', '', clean_desc)
            new_col_name = f"{code}_{clean_desc}".strip('_')
        else:
            new_col_name = re.sub(r'\s+', '_', col_str.strip())
            new_col_name = re.sub(r'[^\w_]+', '', new_col_name).strip('_')
            if not new_col_name: new_col_name = f"col_{hash(col)}"
        count = 1
        original_new_name = new_col_name
        while new_col_name in new_cols.values(): new_col_name = f"{original_new_name}_{count}"; count += 1
        new_cols[col] = new_col_name
    df_cleaned = df_to_clean.rename(columns=new_cols)
    return df_cleaned

# Aplicar limpeza
df = clean_col_names(df)
print("\nNomes das colunas do Survey (AP√ìS LIMPEZA):"); print(df.columns)
df_microdados = clean_col_names(df_microdados)
# if df_microdados is not None: print("\nNomes das colunas do Microdados (AP√ìS LIMPEZA):"); print(df_microdados.columns)


# --- Etapa 2: Sele√ß√£o de Features Relevantes ---
target_column = 'P2_h_Faixa_salarial'
feature_columns = [
    'P1_l_Nivel_de_Ensino','P1_m_√Årea_de_Forma√ß√£o',
    'P2_i_Quanto_tempo_de_experi√™ncia_na_√°rea_de_dados_voc√™_tem','P2_g_Nivel',
    'P4_d_1_SQL', 'P4_d_2_R', 'P4_d_3_Python', 'P4_d_4_CCC', 'P4_d_5_NET', 'P4_d_6_Java',
    'P4_d_7_Julia', 'P4_d_8_SASStata', 'P4_d_9_Visual_BasicVBA', 'P4_d_10_Scala',
    'P4_d_11_Matlab', 'P4_d_12_Rust', 'P4_d_13_PHP', 'P4_d_14_JavaScript',
    'P4_e_Entre_as_linguagens_listadas_abaixo_qual_√©_a_que_voc√™_mais_utiliza_no_trabalho',
]

# Verifica√ß√£o e Sele√ß√£o
print("\nVerificando colunas selecionadas...");
if target_column not in df.columns: raise KeyError(f"Coluna target '{target_column}' ausente.")
print(f"Coluna target encontrada: '{target_column}'")
existing_feature_columns = [col for col in feature_columns if col in df.columns]
missing = [col for col in feature_columns if col not in df.columns]
if missing: print(f"\nAviso: Features n√£o encontradas: {missing}")
if not existing_feature_columns: raise ValueError("Nenhuma feature encontrada.")
df_model = df[existing_feature_columns + [target_column]].copy()
print(f"Shape ap√≥s sele√ß√£o: {df_model.shape}")


# --- Etapa 3: Pr√©-processamento dos Dados ---
# (Mantido integralmente da vers√£o anterior - v3)
# 3.1 Tratar Sal√°rio
def get_salary_midpoint(salary_range):
    # ... (fun√ß√£o da v3) ...
    if pd.isna(salary_range): return np.nan
    s = str(salary_range).lower().replace('r$', '').replace('.', '').replace(',', '.').strip()
    s = s.replace('/m√™s', '').replace('de ', '').replace(' a ', ' ').strip()
    nums = re.findall(r'\d+\.?\d*', s)
    if not nums: return np.nan
    nums = [float(n) for n in nums]
    if 'acima de' in s or 'mais de' in s: return nums[0] * (1.15 if nums[0] >= 25000 else 1.1)
    elif 'menos de' in s or 'at√©' in s: return 800 if nums[0] <= 1000 else nums[0] * 0.9
    elif len(nums) == 2: return (nums[0] + nums[1]) / 2
    elif len(nums) == 1: return nums[0]
    else:
        parts = s.split(); numeric_parts = [float(p) for p in parts if p.replace('.','').isdigit()]
        if len(numeric_parts) == 2: return (numeric_parts[0] + numeric_parts[1]) / 2
        return np.nan

df_model['salary_midpoint'] = df_model[target_column].apply(get_salary_midpoint)
print(f"\nNaNs no sal√°rio midpoint: {df_model['salary_midpoint'].isnull().sum()}")
df_model.dropna(subset=['salary_midpoint'], inplace=True)
if target_column in df_model.columns: df_model.drop(columns=[target_column], inplace=True)
target_column_numeric = 'salary_midpoint'
print(f"Shape ap√≥s tratar sal√°rio: {df_model.shape}")

# 3.2 Tratar Ordinais
ensino_col = 'P1_l_Nivel_de_Ensino'; exp_col = 'P2_i_Quanto_tempo_de_experi√™ncia_na_√°rea_de_dados_voc√™_tem'; senioridade_col = 'P2_g_Nivel'
nivel_ensino_order = ['N√£o tenho gradua√ß√£o formal','Estudante de Gradua√ß√£o','Gradua√ß√£o/Bacharelado','P√≥s-gradua√ß√£o','Mestrado','Doutorado ou Phd']
experiencia_order = ['N√£o tenho experi√™ncia na √°rea de dados','Menos de 1 ano','de 1 a 2 anos','de 3 a 4 anos','de 4 a 6 anos','de 5 a 6 anos','de 7 a 10 anos','Mais de 10 anos']
senioridade_order = ['J√∫nior','Pleno','S√™nior']
ordinal_cols_map = {ensino_col: nivel_ensino_order, exp_col: experiencia_order, senioridade_col: senioridade_order}
processed_ordinal_cols = []; ordinal_encoders = {}

for col, order in ordinal_cols_map.items():
    if col not in df_model.columns: continue
    print(f"Processando ordinal: '{col}'")
    actual_cats = df_model[col].dropna().unique()
    current_order = order # Come√ßa com a ordem base
    extra_cats = [cat for cat in actual_cats if cat not in current_order]
    if extra_cats: print(f"  Aviso: Categorias extras: {extra_cats}"); current_order += extra_cats # Adiciona extras
    final_order = [cat for cat in current_order if cat in actual_cats] # Filtra ordem final
    if not final_order: continue
    ordinal_encoders[col] = {'order': final_order} # Guarda a ordem final
    encoder = OrdinalEncoder(categories=[final_order], handle_unknown='use_encoded_value', unknown_value=np.nan)
    ordinal_encoders[col]['encoder'] = encoder # Guarda o encoder
    try:
        df_model[col] = encoder.fit_transform(df_model[[col]])
        processed_ordinal_cols.append(col)
    except Exception as e: print(f"  Erro encoding {col}: {e}")

ordinal_cols = processed_ordinal_cols
for col in ordinal_cols:
    if col in df_model.columns and df_model[col].isnull().any():
        median_val = df_model[col].median(); df_model[col].fillna(median_val, inplace=True)

# 3.3 Tratar Nominais e Num√©ricas
categorical_cols_nominal = []; numeric_cols = []
remaining_cols = [c for c in existing_feature_columns if c in df_model.columns and c not in ordinal_cols]
for col in remaining_cols:
    if col.startswith('P4_d_'):
         try:
              df_model[col] = pd.to_numeric(df_model[col], errors='coerce')
              if pd.api.types.is_numeric_dtype(df_model[col]): df_model[col].fillna(0, inplace=True); numeric_cols.append(col)
              else: df_model[col]=df_model[col].astype(str).fillna('Desconhecido').astype('category'); categorical_cols_nominal.append(col)
         except: df_model[col]=df_model[col].astype(str).fillna('Desconhecido').astype('category'); categorical_cols_nominal.append(col)
    elif df_model[col].dtype == 'object':
        df_model[col].fillna('Desconhecido', inplace=True); df_model[col] = df_model[col].astype('category'); categorical_cols_nominal.append(col)
    elif pd.api.types.is_numeric_dtype(df_model[col]):
         if df_model[col].isnull().any(): df_model[col].fillna(df_model[col].median(), inplace=True)
         numeric_cols.append(col)
all_numeric_cols = numeric_cols + ordinal_cols
print(f"\nColunas Nominais: {categorical_cols_nominal}")
print(f"Colunas Num√©ricas: {all_numeric_cols}")
# print("\nTipos de dados:"); print(df_model.info()) # Opcional

# 3.4 Verifica√ß√£o Final de NaNs
if df_model.drop(columns=[target_column_numeric]).isnull().sum().sum() == 0: print("\nImputa√ß√£o final OK.")
else: print("\nAten√ß√£o: Ainda h√° NaNs nas features!")


# --- Etapa 4: Divis√£o Treino/Teste ---
X = df_model.drop(columns=[target_column_numeric])
y = df_model[target_column_numeric]
if X.empty or y.empty: raise ValueError("X ou y vazios.")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(f"\nShapes: Treino X={X_train.shape}, Teste X={X_test.shape}")

# Garantir categorias consistentes no teste
categorical_features_lgbm = [col for col in categorical_cols_nominal if col in X_train.columns]
for col in categorical_features_lgbm:
    if col in X_test.columns:
        train_cats = X_train[col].cat.categories
        test_cats = X_test[col].cat.categories
        missing_cats = train_cats.difference(test_cats); extra_cats = test_cats.difference(train_cats)
        if not missing_cats.empty: X_test[col] = X_test[col].cat.add_categories(missing_cats)
        if not extra_cats.empty:
            fill_val = 'Desconhecido' if 'Desconhecido' in train_cats else (X_train[col].mode()[0] if not X_train[col].mode().empty else train_cats[0])
            X_test.loc[X_test[col].isin(extra_cats), col] = fill_val
            try: X_test[col] = X_test[col].cat.remove_categories(extra_cats)
            except: pass # Ignora erro se j√° preenchido
        try: X_test[col] = X_test[col].cat.set_categories(train_cats)
        except ValueError as ve: print(f"  Erro set_categories {col}: {ve}")
print(f"Colunas Categ√≥ricas para LGBM: {categorical_features_lgbm}")


# --- Etapa 5: Indu√ß√£o do Modelo (Instanciar e Treinar) ---
print("\nInstanciando e treinando o modelo LightGBM...")
lgbm = lgb.LGBMRegressor(objective='regression_l1', metric='mae', random_state=42, n_jobs=-1) # Par√¢metros b√°sicos

lgbm_fit_params = {'X': X_train, 'y': y_train, 'eval_set': [(X_test, y_test)], 'eval_metric': 'mae', 'callbacks': [lgb.early_stopping(stopping_rounds=50, verbose=1)]}
if categorical_features_lgbm: lgbm_fit_params['categorical_feature'] = categorical_features_lgbm

try: lgbm.fit(**lgbm_fit_params)
except Exception as e:
     print(f"Erro treino: {e}. Tentando sem 'categorical_feature'...");
     if 'categorical_feature' in lgbm_fit_params: del lgbm_fit_params['categorical_feature']
     try: lgbm.fit(**lgbm_fit_params)
     except Exception as e2: print(f"Erro treino 2: {e2}"); raise e2
print("Treinamento conclu√≠do.")


# --- Etapa 6: Testar o Modelo (Predizer e Avaliar) ---
print("\nPredizendo e avaliando no conjunto de teste...")
y_pred = lgbm.predict(X_test)
mae = mean_absolute_error(y_test, y_pred); rmse = np.sqrt(mean_squared_error(y_test, y_pred)); r2 = r2_score(y_test, y_pred)
print(f"MAE: R$ {mae:,.2f} | RMSE: R$ {rmse:,.2f} | R¬≤: {r2:.4f}")


# --- Etapa 7: Interpreta√ß√£o e Visualiza√ß√£o ---

print("\n--- Interpreta√ß√£o do Modelo ---")

# 7.1 Feature Importance (Geral)
print("\n7.1 Import√¢ncia Geral das Features (Gain):")
num_features = len(X_train.columns); plot_height = max(6, num_features // 3)
plt.figure(figsize=(10, plot_height))
try:
    lgb.plot_importance(lgbm, max_num_features=num_features, importance_type='gain', figsize=(10, plot_height))
    plt.title('Import√¢ncia das Features (LightGBM - Gain)'); plt.tight_layout(); plt.show()
except Exception as e: print(f"Erro plot importance: {e}")

# --- 7.2 Visualiza√ß√£o de √Årvore Individual (Estilo Iris) ---
# NOTA: LightGBM √© um *ensemble*. Visualizar uma √∫nica √°rvore d√° uma ideia,
# mas n√£o representa o comportamento completo do modelo.

print("\n7.2 Visualiza√ß√£o de √Årvore Individual (Exemplo: √Årvore 0):")

# 7.2.1 Usando lgb.plot_tree (Matplotlib)
print("   -> Usando lgb.plot_tree:")
try:
    plt.figure(figsize=(30, 20)) # Tamanho grande para tentar ver detalhes
    lgb.plot_tree(lgbm, tree_index=0, figsize=(30, 20), show_info=['split_gain', 'internal_value', 'internal_count', 'leaf_count'])
    plt.title("Exemplo de √Årvore Individual (√çndice 0) - lgb.plot_tree")
    plt.show()
except Exception as e:
    print(f"   Erro ao plotar com lgb.plot_tree: {e}")

# 7.2.2 Usando graphviz (Mais Detalhado)
print("\n   -> Usando graphviz:")
try:
    booster = lgbm.booster_ # Acessa o modelo booster interno
    graph = lgb.create_tree_digraph(booster, tree_index=0, name=f'Tree {0}')
    display(graph) # Exibe o gr√°fico no Colab/Jupyter
    # Para salvar: graph.render(f"lgbm_tree_{0}", format='png', view=False)
except Exception as e:
    print(f"   Erro ao gerar gr√°fico com graphviz: {e}. Verifique se graphviz est√° instalado.")

# 7.2.3 Usando dtreeviz (Experimental para LGBM)
# print("\n   -> Usando dtreeviz (Experimental):")
# try:
#     # Precisa do booster e dados de treino
#     # ATEN√á√ÉO: Verificar se a vers√£o do dtreeviz suporta LGBM Regressor booster diretamente
#     # Pode ser necess√°rio adaptar ou usar uma interface espec√≠fica se houver
#     viz_model = dtreeviz_model(lgbm.booster_, # Passando o booster
#                                X_train=X_train,
#                                y_train=y_train,
#                                feature_names=X_train.columns.tolist(),
#                                target_name=target_column_numeric,
#                                tree_index=0) # Visualiza a √°rvore 0
#     viz_model.view() # Abre em uma nova janela ou exibe inline
# except ImportError:
#     print("   dtreeviz n√£o instalado ou importado. Pule esta visualiza√ß√£o.")
# except Exception as e:
#     print(f"   Erro ao usar dtreeviz: {e}. Pode n√£o ser totalmente compat√≠vel com LGBMRegressor.")

# --- 7.3 Interpreta√ß√£o com SHAP (Mais Completa para GBM) ---
print("\n--- 7.3 Interpreta√ß√£o SHAP (Recomendada para GBM) ---")
print("Calculando SHAP values...")
try:
    explainer = shap.TreeExplainer(lgbm); shap_values = explainer.shap_values(X_test); X_test_shap = X_test
    print("C√°lculo SHAP conclu√≠do.")

    print("\nPlotando Resumo SHAP (dot)...")
    shap.summary_plot(shap_values, X_test_shap, plot_type="dot", max_display=20, show=False)
    plt.title("SHAP Summary Plot (Dot)"); plt.tight_layout(); plt.show()

    print("\nPlotando Resumo SHAP (bar)...")
    shap.summary_plot(shap_values, X_test_shap, plot_type="bar", max_display=20, show=False)
    plt.title("SHAP Summary Plot (Bar)"); plt.tight_layout(); plt.show()

    # SHAP Dependence Plots (mantidos como antes)
    print("\nPlotando SHAP Dependence Plots...")
    exp_col_shap = 'P2_i_Quanto_tempo_de_experi√™ncia_na_√°rea_de_dados_voc√™_tem' if 'P2_i_Quanto_tempo_de_experi√™ncia_na_√°rea_de_dados_voc√™_tem' in X_test.columns else None
    ensino_col_shap = 'P1_l_Nivel_de_Ensino' if 'P1_l_Nivel_de_Ensino' in X_test.columns else None
    python_col_shap = 'P4_d_3_Python' if 'P4_d_3_Python' in X_test.columns else None
    senioridade_col_shap = 'P2_g_Nivel' if 'P2_g_Nivel' in X_test.columns else None

    def safe_dependence_plot(col_name, interaction_col, shap_vals, data, ordinal_mapping=None):
        # ... (fun√ß√£o da v3) ...
        if col_name and interaction_col and col_name in data.columns and interaction_col in data.columns:
            print(f"\nPlotando Depend√™ncia SHAP: '{col_name}' vs Sal√°rio (colorido por '{interaction_col}')")
            try:
                shap.dependence_plot(col_name, shap_vals, data, interaction_index=interaction_col, show=False)
                plt.title(f"Depend√™ncia: {col_name}\n(Intera√ß√£o: {interaction_col})")
                if ordinal_mapping and col_name in ordinal_mapping and 'encoder' in ordinal_mapping[col_name]:
                     encoder = ordinal_mapping[col_name]['encoder']; cats = ordinal_mapping[col_name]['final_order']
                     try:
                          ticks = encoder.transform([[c] for c in cats]).flatten().tolist()
                          if all(isinstance(t, (int, float)) for t in ticks):
                               vt = sorted(zip(cats, ticks), key=lambda i: i[1]); ft = [i[1] for i in vt]; fl = [i[0] for i in vt]
                               plt.xticks(ft, fl, rotation=45, ha='right', fontsize=8)
                          else: print(f"  Aviso: Ticks inv√°lidos '{col_name}'.")
                     except Exception as tick_e: print(f"  Erro ticks '{col_name}': {tick_e}")
                plt.tight_layout(); plt.show()
            except Exception as e: print(f"  Erro plot depend√™ncia '{col_name}': {e}")
        else: print(f"N√£o plotado: '{col_name}' ou '{interaction_col}' n√£o encontrado.")

    safe_dependence_plot(exp_col_shap, ensino_col_shap, shap_values, X_test_shap, ordinal_encoders)
    safe_dependence_plot(python_col_shap, exp_col_shap, shap_values, X_test_shap, ordinal_encoders)
    safe_dependence_plot(ensino_col_shap, exp_col_shap, shap_values, X_test_shap, ordinal_encoders)
    safe_dependence_plot(senioridade_col_shap, exp_col_shap, shap_values, X_test_shap, ordinal_encoders)

except Exception as e:
    print(f"Erro no c√°lculo ou plotagem SHAP: {e}")

print("\n--- Fim da An√°lise ---")
```
