```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from mpl_toolkits.mplot3d import Axes3D # Para gráficos 3D com matplotlib
import os # Para criar diretórios

# --- Configurações e Criação de Pasta para Gráficos ---
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
pd.set_option('display.max_columns', None)

# Criar pasta para salvar os gráficos, se não existir
output_dir = 'graficos_analise_exploratoria'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Diretório '{output_dir}' criado para salvar os gráficos.")
else:
    print(f"Diretório '{output_dir}' já existe. Gráficos serão salvos nele.")

# Função auxiliar para salvar gráficos matplotlib/seaborn
def save_plt_figure(filename_base, tight_layout=True):
    if tight_layout:
        plt.tight_layout()
    filepath = os.path.join(output_dir, f"{filename_base}.png")
    plt.savefig(filepath)
    print(f"Gráfico salvo em: {filepath}")
    plt.show() # Continuar mostrando o gráfico no notebook

# Função auxiliar para salvar gráficos Plotly
def save_plotly_figure(fig, filename_base):
    # Salvar como HTML
    filepath_html = os.path.join(output_dir, f"{filename_base}.html")
    fig.write_html(filepath_html)
    print(f"Gráfico interativo salvo em: {filepath_html}")

    # Tentar salvar como PNG (requer kaleido)
    try:
        filepath_png = os.path.join(output_dir, f"{filename_base}.png")
        fig.write_image(filepath_png, scale=2) # scale para melhor resolução
        print(f"Gráfico estático salvo em: {filepath_png}")
    except ValueError as e:
        print(f"Não foi possível salvar {filename_base}.png (Plotly). Erro: {e}")
        print("Certifique-se de ter a biblioteca 'kaleido' instalada: pip install kaleido")
    fig.show() # Continuar mostrando o gráfico no notebook


# --- 1. Carregar e Inspecionar Dados ---
print("\n--- 1. Carregando e Inspecionando Dados ---")
try:
    df_profissionais = pd.read_csv('/kaggle/input/dataset-clean/dados_limpos.csv') # [2]
    df_microdados_uf = pd.read_csv('/kaggle/input/dataset-microdados-agrupados-por-uf-cleaned/microdados_agrupados_por_uf.csv') # [1]
except FileNotFoundError:
    print("Erro: Um ou ambos os arquivos CSV não foram encontrados. Verifique os caminhos.")
    exit()

print("\n### Dataset de Profissionais de Dados (dados_limpos.csv):")
print("Primeiras 5 linhas:")
print(df_profissionais.head())
print("\nInformações gerais:")
df_profissionais.info()

print("\n### Dataset de Microdados por UF (microdados_agrupados_por_uf.csv):")
print("Primeiras 5 linhas:")
print(df_microdados_uf.head())
print("\nInformações gerais:")
df_microdados_uf.info()


# --- 2. Pré-processamento de df_profissionais ---
print("\n--- 2. Pré-processamento do Dataset de Profissionais ---")
df_profissionais_proc = df_profissionais.copy()

faixa_salarial_map = {
    'Menos de R$ 1.000/mês': 500, 'de R$ 1.001/mês a R$ 2.000/mês': 1500,
    'de R$ 2.001/mês a R$ 3.000/mês': 2500, 'de R$ 3.001/mês a R$ 4.000/mês': 3500,
    'de R$ 4.001/mês a R$ 6.000/mês': 5000, 'de R$ 6.001/mês a R$ 8.000/mês': 7000,
    'de R$ 8.001/mês a R$ 12.000/mês': 10000, 'de R$ 12.001/mês a R$ 16.000/mês': 14000,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000, 'de R$ 20.001/mês a R$ 25.000/mês': 22500,
    'de R$ 25.001/mês a R$ 30.000/mês': 27500, 'de R$ 30.001/mês a R$ 40.000/mês': 35000,
    'Acima de R$ 40.001/mês': 45000
}
df_profissionais_proc['Salario_Estimado'] = df_profissionais_proc['Faixa salarial mensal'].map(faixa_salarial_map)
ordem_faixa_salarial = list(faixa_salarial_map.keys())
df_profissionais_proc['Faixa salarial mensal'] = pd.Categorical(
    df_profissionais_proc['Faixa salarial mensal'], categories=ordem_faixa_salarial, ordered=True
)

experiencia_map = {
    'Menos de 1 ano': 0.5, 'de 1 a 2 anos': 1.5, 'de 3 a 4 anos': 3.5,
    'de 5 a 6 anos': 5.5, 'de 7 a 10 anos': 8.5, 'Mais de 10 anos': 12
}
df_profissionais_proc['Experiencia_Anos_Estimados'] = df_profissionais_proc['Tempo de experiência na área de dados'].map(experiencia_map)
ordem_experiencia = list(experiencia_map.keys())
df_profissionais_proc['Tempo de experiência na área de dados'] = pd.Categorical(
    df_profissionais_proc['Tempo de experiência na área de dados'], categories=ordem_experiencia, ordered=True
)

ordem_nivel_ensino = [
    'Estudante de Graduação', 'Graduação/Bacharelado',
    'Pós-graduação', 'Mestrado', 'Doutorado ou Phd'
]
nivel_ensino_map_num = { nivel: i for i, nivel in enumerate(ordem_nivel_ensino) }
df_profissionais_proc['Nivel_Ensino_Ordinal'] = df_profissionais_proc['Nível de ensino alcançado'].map(nivel_ensino_map_num)
df_profissionais_proc['Nível de ensino alcançado'] = pd.Categorical(
    df_profissionais_proc['Nível de ensino alcançado'], categories=ordem_nivel_ensino, ordered=True
)

print("\nDataset de profissionais após pré-processamento (novas colunas e tipos):")
print(df_profissionais_proc[['Faixa salarial mensal', 'Salario_Estimado',
                             'Tempo de experiência na área de dados', 'Experiencia_Anos_Estimados',
                             'Nível de ensino alcançado', 'Nivel_Ensino_Ordinal']].head())

# --- 3. Análise Exploratória Univariada (em df_profissionais_proc) ---
print("\n--- 3. Análise Exploratória Univariada ---")

# 3.1. Distribuição da Faixa Salarial (Categórica)
plt.figure(figsize=(14, 7))
sns.countplot(data=df_profissionais_proc, y='Faixa salarial mensal', order=ordem_faixa_salarial, palette='viridis')
plt.title('Distribuição de Profissionais por Faixa Salarial Mensal')
plt.xlabel('Contagem')
plt.ylabel('Faixa Salarial Mensal')
save_plt_figure('distribuicao_faixa_salarial')

# 3.2. Distribuição do Salário Estimado (Numérica)
plt.figure(figsize=(10, 6))
sns.histplot(df_profissionais_proc['Salario_Estimado'].dropna(), kde=True, bins=15) # Adicionado dropna()
plt.title('Distribuição do Salário Estimado (R$)')
plt.xlabel('Salário Estimado (R$)')
plt.ylabel('Frequência')
save_plt_figure('distribuicao_salario_estimado')

# 3.3. Distribuição do Nível de Ensino
plt.figure(figsize=(10, 6))
sns.countplot(data=df_profissionais_proc, y='Nível de ensino alcançado', order=ordem_nivel_ensino, palette='crest')
plt.title('Distribuição de Profissionais por Nível de Ensino')
plt.xlabel('Contagem')
plt.ylabel('Nível de Ensino')
save_plt_figure('distribuicao_nivel_ensino')

# 3.4. Distribuição da Área de Formação Acadêmica
plt.figure(figsize=(12, 8))
sns.countplot(data=df_profissionais_proc, y='Área de formação acadêmica',
              order=df_profissionais_proc['Área de formação acadêmica'].value_counts().index, palette='flare')
plt.title('Distribuição de Profissionais por Área de Formação Acadêmica')
plt.xlabel('Contagem')
plt.ylabel('Área de Formação')
save_plt_figure('distribuicao_area_formacao')

# 3.5. Distribuição do Tempo de Experiência
plt.figure(figsize=(10, 6))
sns.countplot(data=df_profissionais_proc, y='Tempo de experiência na área de dados', order=ordem_experiencia, palette='magma')
plt.title('Distribuição de Profissionais por Tempo de Experiência em Dados')
plt.xlabel('Contagem')
plt.ylabel('Tempo de Experiência')
save_plt_figure('distribuicao_tempo_experiencia')

# 3.6. Distribuição por UF onde mora (Top 10)
plt.figure(figsize=(12, 7))
top_10_uf = df_profissionais_proc['UF onde mora'].value_counts().nlargest(10)
sns.barplot(x=top_10_uf.index, y=top_10_uf.values, palette='Blues_r')
plt.title('Top 10 UF de Residência dos Profissionais de Dados')
plt.xlabel('UF')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
save_plt_figure('distribuicao_top10_uf')


# --- 4. Análise Exploratória Bivariada (em df_profissionais_proc) ---
print("\n--- 4. Análise Exploratória Bivariada ---")

# 4.1. Formação Acadêmica vs. Salário Estimado (Matplotlib/Seaborn)
plt.figure(figsize=(12, 7))
sns.boxplot(data=df_profissionais_proc, x='Salario_Estimado', y='Nível de ensino alcançado',
            order=ordem_nivel_ensino, palette='crest')
plt.title('Salário Estimado por Nível de Ensino')
plt.xlabel('Salário Estimado (R$)')
plt.ylabel('Nível de Ensino')
save_plt_figure('boxplot_salario_por_nivel_ensino_seaborn')

# 4.1. Formação Acadêmica vs. Salário Estimado (Plotly)
fig_box_sal_nivel = px.box(df_profissionais_proc.dropna(subset=['Salario_Estimado', 'Nível de ensino alcançado']), # Adicionado dropna
                           x='Salario_Estimado', y='Nível de ensino alcançado',
                           category_orders={'Nível de ensino alcançado': ordem_nivel_ensino},
                           title='Salário Estimado por Nível de Ensino (Plotly)',
                           labels={'Salario_Estimado': 'Salário Estimado (R$)', 'Nível de ensino alcançado': 'Nível de Ensino'},
                           color='Nível de ensino alcançado')
save_plotly_figure(fig_box_sal_nivel, 'boxplot_salario_por_nivel_ensino_plotly')


# 4.2. Área de Formação vs. Salário Estimado (Top 5 áreas)
top_5_areas = df_profissionais_proc['Área de formação acadêmica'].value_counts().nlargest(5).index
df_top_areas = df_profissionais_proc[df_profissionais_proc['Área de formação acadêmica'].isin(top_5_areas)]
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_top_areas, x='Salario_Estimado', y='Área de formação acadêmica',
            order=top_5_areas, palette='flare')
plt.title('Salário Estimado por Área de Formação (Top 5)')
plt.xlabel('Salário Estimado (R$)')
plt.ylabel('Área de Formação')
save_plt_figure('boxplot_salario_por_area_formacao_top5')

# 4.3. Experiência Profissional vs. Salário Estimado (Matplotlib/Seaborn)
plt.figure(figsize=(12, 7))
sns.boxplot(data=df_profissionais_proc, x='Salario_Estimado', y='Tempo de experiência na área de dados',
            order=ordem_experiencia, palette='magma')
plt.title('Salário Estimado por Tempo de Experiência em Dados')
plt.xlabel('Salário Estimado (R$)')
plt.ylabel('Tempo de Experiência')
save_plt_figure('boxplot_salario_por_experiencia_seaborn')

# 4.3. Experiência Profissional vs. Salário Estimado (Plotly)
fig_box_sal_exp = px.box(df_profissionais_proc.dropna(subset=['Salario_Estimado', 'Tempo de experiência na área de dados']), # Adicionado dropna
                         x='Salario_Estimado', y='Tempo de experiência na área de dados',
                         category_orders={'Tempo de experiência na área de dados': ordem_experiencia},
                         title='Salário Estimado por Tempo de Experiência (Plotly)',
                         labels={'Salario_Estimado': 'Salário Estimado (R$)', 'Tempo de experiência na área de dados': 'Tempo de Experiência'},
                         color='Tempo de experiência na área de dados')
save_plotly_figure(fig_box_sal_exp, 'boxplot_salario_por_experiencia_plotly')

# 4.4. Heatmap de Correlação
cols_numericas_interesse = ['Salario_Estimado', 'Experiencia_Anos_Estimados', 'Nivel_Ensino_Ordinal']
# Remover linhas com NaN em qualquer uma dessas colunas para a correlação
df_corr = df_profissionais_proc[cols_numericas_interesse].dropna()
matriz_corr = df_corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlação entre Salário, Experiência (anos) e Nível de Ensino (ordinal)')
save_plt_figure('heatmap_correlacao_salario_exp_ensino', tight_layout=False) # tight_layout=False às vezes é melhor para heatmaps


# --- 5. Análise Exploratória Multivariada (Interações em df_profissionais_proc) ---
print("\n--- 5. Análise Exploratória Multivariada (Interações) ---")

# 5.1. Interação: Salário Estimado por Experiência, segmentado por Nível de Ensino
plt.figure(figsize=(14, 8))
# Remover NaNs para evitar problemas com lineplot e ci
df_lineplot = df_profissionais_proc.dropna(subset=['Experiencia_Anos_Estimados', 'Salario_Estimado', 'Nível de ensino alcançado'])
sns.lineplot(data=df_lineplot, x='Experiencia_Anos_Estimados', y='Salario_Estimado',
             hue='Nível de ensino alcançado', hue_order=ordem_nivel_ensino,
             palette='viridis', marker='o', errorbar=('ci', 95))
plt.title('Salário Médio Estimado vs. Anos de Experiência por Nível de Ensino')
plt.xlabel('Anos de Experiência Estimados')
plt.ylabel('Salário Médio Estimado (R$)')
plt.legend(title='Nível de Ensino')
plt.grid(True)
save_plt_figure('lineplot_salario_exp_por_nivel_ensino')

# 5.2. Boxplot: Salário por Experiência, facetado por Nível de Ensino
df_catplot = df_profissionais_proc.dropna(subset=['Tempo de experiência na área de dados', 'Salario_Estimado', 'Nível de ensino alcançado'])
g = sns.catplot(data=df_catplot, x='Tempo de experiência na área de dados', y='Salario_Estimado',
                col='Nível de ensino alcançado', col_wrap=3, kind='box',
                order=ordem_experiencia, col_order=ordem_nivel_ensino, palette='Paired', height=5, aspect=1.2)
g.set_titles("{col_name}")
g.set_xticklabels(rotation=30, ha='right')
g.set_ylabels('Salário Estimado (R$)')
g.set_xlabels('Tempo de Experiência')
plt.suptitle('Salário Estimado por Experiência, Agrupado por Nível de Ensino', y=1.03)
# Salvar catplot é um pouco diferente, pois g é um FacetGrid
filepath_catplot = os.path.join(output_dir, "catplot_salario_exp_facet_nivel_ensino.png")
g.savefig(filepath_catplot)
print(f"Gráfico salvo em: {filepath_catplot}")
plt.show()


# 5.3. Gráfico 3D: Salário vs. Experiência Numérica vs. Nível de Ensino Numérico (Plotly)
df_plot3d = df_profissionais_proc.dropna(subset=['Salario_Estimado', 'Experiencia_Anos_Estimados', 'Nivel_Ensino_Ordinal', 'Nível de ensino alcançado'])
fig_3d_scatter = px.scatter_3d(df_plot3d,
                               x='Experiencia_Anos_Estimados',
                               y='Nivel_Ensino_Ordinal',
                               z='Salario_Estimado',
                               color='Nível de ensino alcançado',
                               category_orders={'Nível de ensino alcançado': ordem_nivel_ensino},
                               labels={
                                   'Experiencia_Anos_Estimados': 'Experiência (Anos Estimados)',
                                   'Nivel_Ensino_Ordinal': 'Nível de Ensino (Ordinal)',
                                   'Salario_Estimado': 'Salário Estimado (R$)'
                               },
                               title='Interação 3D: Salário, Experiência e Nível de Ensino')
fig_3d_scatter.update_layout(scene=dict(
                    yaxis = dict(
                         tickmode = 'array',
                         tickvals = list(nivel_ensino_map_num.values()),
                         ticktext = list(nivel_ensino_map_num.keys())
                    )))
save_plotly_figure(fig_3d_scatter, 'scatter3d_salario_exp_ensino')

# 5.4 Gráfico de Superfície 3D (Plotly)
df_agg_3d = df_profissionais_proc.groupby(['Experiencia_Anos_Estimados', 'Nivel_Ensino_Ordinal'], as_index=False)['Salario_Estimado'].mean()
df_agg_3d = df_agg_3d.dropna()

if not df_agg_3d.empty and len(df_agg_3d['Experiencia_Anos_Estimados'].unique()) > 1 and len(df_agg_3d['Nivel_Ensino_Ordinal'].unique()) > 1:
    try:
        pivot_3d = df_agg_3d.pivot(index='Nivel_Ensino_Ordinal', columns='Experiencia_Anos_Estimados', values='Salario_Estimado')
        pivot_3d = pivot_3d.fillna(0) # Preencher NaNs para o plot

        X_surf = pivot_3d.columns.values
        Y_surf = pivot_3d.index.values
        Z_surf = pivot_3d.values

        fig_3d_surface = go.Figure(data=[go.Surface(z=Z_surf, x=X_surf, y=Y_surf, colorscale='Viridis')])
        fig_3d_surface.update_layout(title='Superfície 3D: Salário Médio por Experiência e Nível de Ensino',
                                     scene=dict(
                                         xaxis_title='Experiência (Anos Estimados)',
                                         yaxis_title='Nível de Ensino (Ordinal)',
                                         zaxis_title='Salário Médio Estimado (R$)',
                                         yaxis = dict(
                                             tickmode = 'array',
                                             tickvals = list(nivel_ensino_map_num.values()),
                                             ticktext = list(nivel_ensino_map_num.keys())
                                         )
                                     ))
        save_plotly_figure(fig_3d_surface, 'surface3d_salario_medio_exp_ensino')
    except Exception as e:
        print(f"Não foi possível gerar o gráfico de superfície 3D: {e}")
else:
    print("Dados insuficientes ou não variados o suficiente para gerar gráfico de superfície 3D após agregação.")


# --- 6. Integração dos Dados ---
print("\n--- 6. Integração dos Datasets ---")
df_integrado = pd.merge(df_profissionais_proc, df_microdados_uf,
                        left_on='UF onde mora', right_on='UF',
                        how='left')

print("\nDataset integrado (primeiras linhas):")
print(df_integrado.head())


# --- 7. Análise Exploratória com Dados Integrados ---
print("\n--- 7. Análise Exploratória com Dados Integrados ---")
if df_integrado['Docentes_Doutorado'].isnull().all():
    print("Merge não resultou em dados úteis de microdados para os profissionais.")
else:
    df_integrado['Total_Docentes_UF'] = df_integrado[['Docentes_Graduacao', 'Docentes_Especializacao',
                                                  'Docentes_Mestrado', 'Docentes_Doutorado']].sum(axis=1)
    df_integrado['Prop_Docentes_Doutorado_UF'] = df_integrado['Docentes_Doutorado'] / df_integrado['Total_Docentes_UF']
    df_integrado['Prop_Docentes_Doutorado_UF'].replace([np.inf, -np.inf], np.nan, inplace=True) # Tratar divisão por zero se Total_Docentes_UF for 0

    # 7.1. Salário Estimado vs. Proporção de Docentes com Doutorado na UF
    df_plot_integrado = df_integrado.dropna(subset=['Salario_Estimado', 'Prop_Docentes_Doutorado_UF', 'Nível de ensino alcançado', 'Experiencia_Anos_Estimados'])
    if not df_plot_integrado.empty:
        plt.figure(figsize=(14, 9)) # Ajustado tamanho
        sns.scatterplot(data=df_plot_integrado, x='Prop_Docentes_Doutorado_UF', y='Salario_Estimado',
                        hue='Nível de ensino alcançado',
                        size='Experiencia_Anos_Estimados',
                        palette='Set2', alpha=0.7, sizes=(20, 200)) # Ajustado sizes
        plt.title('Salário Estimado vs. Proporção de Docentes com Doutorado na UF de Residência')
        plt.xlabel('Proporção de Docentes com Doutorado na UF de Residência')
        plt.ylabel('Salário Estimado (R$)')
        plt.legend(title='Legenda', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) # Ajustado legenda
        save_plt_figure('scatterplot_salario_vs_prop_doc_doutorado_uf')

        if 'Prop_Docentes_Doutorado_UF' in df_plot_integrado.columns and df_plot_integrado['Prop_Docentes_Doutorado_UF'].nunique() > 1:
             corr_sal_prop_dout = df_plot_integrado[['Salario_Estimado', 'Prop_Docentes_Doutorado_UF']].corr().iloc[0,1]
             print(f"\nCorrelação entre Salário Estimado e Proporção de Docentes com Doutorado na UF: {corr_sal_prop_dout:.2f}")
    else:
        print("Não há dados suficientes após filtrar NaNs para o gráfico Salário vs. Prop. Docentes Doutorado.")

    # 7.2. Média Salarial por UF e comparação com total de docentes na UF
    # Remover NaNs antes do groupby para evitar problemas
    df_integrado_agg = df_integrado.dropna(subset=['UF onde mora', 'Salario_Estimado', 'Total_Docentes_UF'])
    media_salarial_uf = df_integrado_agg.groupby('UF onde mora')['Salario_Estimado'].mean().sort_values(ascending=False)
    total_docentes_uf_agg = df_integrado_agg.groupby('UF onde mora')['Total_Docentes_UF'].first().sort_values(ascending=False)

    if not media_salarial_uf.empty and not total_docentes_uf_agg.empty:
        df_comparacao_uf = pd.DataFrame({'Media_Salarial': media_salarial_uf, 'Total_Docentes': total_docentes_uf_agg}).dropna().sort_values(by='Media_Salarial', ascending=False)

        if not df_comparacao_uf.empty:
            fig, ax1 = plt.subplots(figsize=(16, 9)) # Ajustado tamanho
            color = 'tab:red'
            ax1.set_xlabel('UF onde mora')
            ax1.set_ylabel('Salário Médio Estimado (R$)', color=color)
            ax1.bar(df_comparacao_uf.index, df_comparacao_uf['Media_Salarial'], color=color, alpha=0.6, label='Salário Médio')
            ax1.tick_params(axis='y', labelcolor=color)
            ax1.tick_params(axis='x', rotation=90)
            ax1.grid(False) # Remover grid do ax1 para clareza

            ax2 = ax1.twinx()
            color = 'tab:blue'
            ax2.set_ylabel('Total de Docentes na UF', color=color)
            ax2.plot(df_comparacao_uf.index, df_comparacao_uf['Total_Docentes'], color=color, marker='o', linestyle='--', label='Total Docentes')
            ax2.tick_params(axis='y', labelcolor=color)
            ax2.grid(False) # Remover grid do ax2

            fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes) # Adicionar legenda
            plt.title('Salário Médio Estimado e Total de Docentes por UF')
            save_plt_figure('bar_line_salario_medio_total_docentes_uf')
        else:
            print("DataFrame df_comparacao_uf está vazio após o merge/dropna.")
    else:
        print("Não foi possível gerar o gráfico comparativo por UF (dados agregados insuficientes).")

print("\n--- Análise Exploratória Concluída ---")
print(f"Todos os gráficos foram salvos no diretório: {output_dir}")

```
